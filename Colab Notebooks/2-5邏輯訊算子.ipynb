```python
num = eval(input('Enter a number: '))
if (85 <= num) and (num <= 95):
  print('%d is in the betwween 85 and 95' % (num))
else:
  if (num <= 84) or (num >= 96):
    print('%d is not in the betwween 85 and 95' % (num))
print('over')
```
