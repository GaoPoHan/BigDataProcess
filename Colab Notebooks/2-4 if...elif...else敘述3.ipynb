```python
height = eval(input('Enter height in centermeters: '))
weight = eval(input('Enter weight in kliograms'))

bmi = weight / (height/100)**2
print('Your BMI is %6.2f'%(bmi))

if bmi < 18.5:
  print('Underweight')
elif bmi < 25.0:
  print('Normal')
elif bmi < 30:
  print('Overweight')
else:
  print('Obese')
```
